```python
import seaborn as sns
fmri =sns.load_dataset("fmri")
sns.pairplot(fmri)
sns.jointplot(x=fmri.timepoint ,y= fmri.signal)
sns.catplot(x='timepoint',y='signal',data = fmri,color ='r')
sns.relplot(x =fmri.timepoint , y = fmri.event , data = fmri , style = 'event',hue='timepoint')
sns.displot(fmri['timepoint'])
```
